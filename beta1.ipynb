```python
new_df = raw_data.copy()
new_df
```

```python
# dropping the ID column as it have no hidden or predictive power 
new_df = new_df.drop(['ID'],axis = 1)
new_df.head()
```

```python
# working on reason of absense
reason_absense = new_df['Reason for Absence']
```

```python
#changing different reason to dummy variables
alpha = pd.get_dummies(reason_absense)
alpha['check'] = alpha.sum(axis = 1)
alpha
```

```python
alpha['check'].unique()
alpha['check'].sum(axis = 0)
```

```python
alpha.columns.values
```

```python
alpha = alpha.drop([0],axis = 1)# we drop 0 for  multicolinearity
```

```python
# grouping the data set in different 4 group to reduce variables 
alpha.iloc[:,1:14]
```

```python
group_1 = alpha.iloc[:,1:14].max(axis = 1)
group_2   = alpha.iloc[:,15:17].max(axis = 1)
group_3 = alpha.iloc[:,18:21].max(axis = 1)
group_4 = alpha.iloc[:,22:].max(axis = 1)
```

```python
new_df = pd.concat([new_df,group_1,group_2,group_3,group_4],axis = 1)
new_df = new_df.drop([0],axis = 1)
new_df = new_df.drop([1],axis = 1)
new_df = new_df.drop([2],axis = 1)
new_df = new_df.drop([3],axis = 1)
```

```python
save_df['Date'].shape[0]
```

```python
for i in range (save_df['Date'].shape[0]):
    month_list.append(save_df['Date'][i].month)
```

```python
save_df['Month'] = month_list
save_df.head()# note: in .month 1 t0 12  has their own meaning as it is 
```

```python
# now extracting the values of the weekdays so that we can get the day  of week
# the week days are from 0 to 6 hence total 7 weekdays
def to_weekday(date):
    return date.weekday()
```

```python
save_df['Day of week'] = save_df['Date'].apply(to_weekday)
save_df.head()
```
